## Escalado (metodo de estandarización)

Algunos de los metodos de machine learning necesita que los datos estén escalados

```python
x_scaled = preprocessing.scale(df)
x_scaled
```

```python
x_scaled.mean(axis=0)#Media de cada una de las columnas
```

```python
x_scaled.std(axis=0)#Calcula desviacion estandar
```

# k-Means, el algoritmo de clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/ordonez/Documents/UCASAL/notebooks/2008.csv",nrows=1e5)
newdf = df[["AirTime","DepDelay"]].dropna() #Eliminamos los casos faltantes de estos 2 col.
```

### Entrenamineto de un modelo de kmeans

```python
kmeans = KMeans(n_clusters=4,random_state=0).fit(newdf)
kmeans.labels_
```

```python
#Usamos la siguiente para ver como han quedado distribuidos los grupos
np.unique(kmeans.labels_,return_counts=True)
```

### Visualizamos gráficamente como quedan distribuidos los centroides

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c= kmeans.labels_)
plt.show()
```

### muestra donde quedan situados los centroides

```python
kmeans.cluster_centers_
```

### Usamos el objeto kmeans para predecir nuevos puntos a que grupo van a pertencer

```python
#En lugar de cargar 100.000 puntos vamos a cargar un 1.000.000
alldf = pd.read_csv("C:/Users/ordonez/Documents/UCASAL/notebooks/2008.csv",nrows=1e6)[["AirTime","DepDelay"]].dropna()
```

```python
kmeans.predict(alldf)[0:50] #como clasifica
```

### Ejecutamos el metodos con mas de 2 variables

```python
#Creamos un dataframe con 5 variables
newdf = df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
kmeans = KMeans(n_clusters=4,random_state=0).fit(newdf)
kmeans.labels_
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c= kmeans.labels_)
plt.show()
```

### Probando con mas cluster

```python
kmeans = KMeans(n_clusters=6,random_state=0).fit(newdf)
kmeans.labels_
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c= kmeans.labels_)
plt.show()
```

# El algoritmo hierarchical clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/ordonez/Documents/UCASAL/notebooks/2008.csv",nrows=1e4)
newdf = df[["AirTime","DepDelay"]].dropna()
```

## Ajustamos el modelo

```python
clstr = AgglomerativeClustering(n_clusters=5)
clstr.fit(newdf)
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c=clstr.fit_predict(newdf))
plt.show()
```

## Ajustando el modelo con muchas variables

```python
newdf = df[["AirTime","Distance","TaxiOut","ArrDelay","DepDelay"]].dropna()
```

```python
clstr = AgglomerativeClustering(n_clusters=20)
clstr.fit(newdf)
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c=clstr.fit_predict(newdf))
plt.show()
```
