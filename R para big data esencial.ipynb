## Importando un archivo .csv a un dataframe

```r
library(readr)
df <- as.data.frame(read_csv("2008.csv"))
```

```r
df
```

```r
#Guardo una columna en delay
delay <- df$ArrDelay
delay <- delay[!is.na(delay)]
```

## Calculo del tiempo de ejecución

```r
start <- Sys.time()
mean(delay)
print(Sys.time() - start)
```

```r
# para evaluar una función
system.time(median(delay))
```

## rbenchmark

```r
library(rbenchmark)
```

```r
benchmark(
    "Media funcion" = {
        mean(delay)
    },
    "Media con funciones" = {
        sum(delay)/nrow(delay)
    },
    "Media manual" = {
        media <- 0
        for(i in 1:nrow(df)){
            media <- media + delay[i]
        }
        media/nrow(df)
    },
    replications = 20,
    columns = c("test","replications",
                "elapsed","relative")
)
```

# Programación vectorial en R

```r
df <- as.data.frame(read_csv("2008.csv", n_max=100000))
delay <- df$ArrDelay
delay <- delay[!is.na(delay)] #Limpiamos de casos faltantes
```

```r
delay
```

## Suma elemento a elemento de dos columnas de un dataframe

```r
df$TaxiIn + df$TaxiOut
```

```r
df$TaxiIn > df$TaxiOut
```

```r
delay[delay > 0]
```

```r
delay[delay > 0] * 2 # multiplica elemneto a elemento por 2
```

```r
a <- delay[delay > 0] * 2 
b <- delay[delay > 0] + delay[delay > 0]
a == b
```

```r
all(a == b)
```

## Producto matricial %*%

```r
delay %*% rep(1,length(delay))
```

```r
mean(delay)
```

```r
delay %*% rep(1,length(delay)) / length(delay)
```

```r
rep(1,length(delay))
```

```r
df2 <- df[,c("CarrierDelay","WeatherDelay",
            "NASDelay","SecurityDelay",
            "LateAircraftDelay")]
df2 <- df2[complete.cases(df2),]
```

```r
df2 <- as.matrix(df2)
```

```r
df2 %*% rep(1,ncol(df2))
```

```r
rep(1,nrow(df2)) %*% df2
```

```r
resultado <- rep(1,nrow(df2)) %*% df2
```

```r
resultado[5] == sum(df2[,5])
```
